#### Bar charts

This type of chart is very good with categorical data to depict the summary of the information that you have.

As an example we will explore the total quantity of goods per city

```python

plt.bar(data=market_sales_df,x="City",height="Quantity")
plt.show()
```

```python
x = market_sales_df["City"]
y = market_sales_df["Quantity"]
plt.bar(x=x,height=y)
plt.show()
```

```python
# adding more context to the bar chart
x = market_sales_df["City"]
y = market_sales_df["Quantity"]
plt.bar(x=x,height=y)

# adding labels to the chart
plt.ylabel("Quantity")
plt.xlabel("City")
plt.title("Max Quantity per City")
plt.show()
```

#### Groupby 

Similar to Pivot chart in Excel and summarises the whole dataset based on a particular column

```python
market_sales_df.head()
```

```python
# basic implementation of groupby
market_sales_df.groupby("City")
```

```python
market_sales_df.groupby("City").describe(include="all")
```

```python
market_sales_df.groupby("City")["Quantity"]
```

Describe is not the only function you can use with a groupby. Others include:
1. Count
2. Sum
3. Average
4. Min
5. Max

```python
market_sales_df.groupby("City")["Quantity"].sum()
```

```python
market_sales_df.groupby("City")["Total"].sum()
```

```python
market_sales_df.groupby(["City","Branch"])["Quantity"].sum()
```

```python
city_one = market_sales_df[market_sales_df["City"] == "Yangon"]
city_one["Branch"].unique()
```

```python
market_sales_df.groupby(["City","Gender"])["Quantity"].sum()
```

```python
city_quantity = market_sales_df.groupby("City")["Quantity"].sum().to_frame()
```

```python
type(city_quantity)
```

```python
plt.bar(data=city_quantity,x=city_quantity.index,height="Quantity")
plt.show()
```

```python
plt.bar(data=city_quantity,x=city_quantity.index,height="Quantity")
plt.xlabel("City")
plt.ylabel("Quantity")
plt.title("Total Quantity sold per City")
plt.show()
```

```python
customers = market_sales_df["Date"].value_counts().to_frame().sort_index()
```

```python
# this sets the size of your chart
plt.figure(figsize=(20,10))

plt.plot(customers.index,customers.Date)

# adjusts the x-axis, rotating the x-values as they are printed
plt.xticks(rotation=90)

# labels
plt.xlabel("Date")
plt.ylabel("Customer Count")
plt.title("Total Customers per Date")

plt.show()
```

```python
time = market_sales_df.groupby("Date")["Quantity"].sum().to_frame()
time
```

```python
plt.plot(time.index,time.Quantity)
plt.show()
```

```python
# this sets the size of your chart
plt.figure(figsize=(20,10))

plt.plot(time.index,time.Quantity)

# adjusts the x-axis, rotating the x-values as they are printed
plt.xticks(rotation=90)

# labels
plt.xlabel("Date")
plt.ylabel("Quantity")
plt.title("Total Quantity per Date")

plt.show()
```

```python
month = market_sales_df.groupby("Month")["Quantity"].sum().to_frame()
```

```python
month
```

```python
plt.bar(data=month,x=month.index,height="Quantity")
plt.xlabel("Month")
plt.ylabel("Quantity")
plt.title("Total Quantity sold per Month")
plt.show()
```

### if, elif and else statement

```python
x = 30

if x > 0:
    print("the number is greater than zero")
```

```python
if x > 50:
    print("this number is greater than fifty")
```

```python
if x > 50:
    print("this number is greater than fifty")
else:
    print("this number is not greater than fifty")
```

```python
if x > 50:
    print("this number if greater than fifty")
elif x > 20:
    print("ok this number is actually greater than twenty")
else:
    print("this number does not meet the criteria")
```

```python
x = 20
```

```python
if x > 50:
    print("this number if greater than fifty")
elif x == 50:
    print("this number is 50")
elif x >= 20:
    print("this number is greater than 20")
else:
    print("criteria not met")
```

### For statement

```python
weekdays = ["Monday","Tuesday","Wednesday","Thursday","Friday"]
weekends = ["Saturday","Sunday"]
days = weekdays + weekends
```

```python
days
```

```python
for d in days:
    print(d)
```

```python
for letter in "this is an example":
    print(letter)
```

```python
for y in range(10):
    print(y)
```

### While statement

```python
x > 5

while x > 0:
    print("to infinity")
```

```python
x = 5

while x > 0:
    print("testing " + str(x))
    x = x - 1
```
